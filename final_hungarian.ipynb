```python
#step 1: find smallest element in each row of cost matrix and subtract it from each element in the row 

def step_1(input_matrix):
    reduced_matrix = []
    for i in range(len(input_matrix)): 
        new_row=[]
        row = input_matrix[i]
        smallest_element = 10000
        for k in range(len(row)): 
            if row[k] < smallest_element: 
                smallest_element = row[k]
        for j in range(len(row)): 
            new_row.append(row[j] - smallest_element)
        reduced_matrix.append(new_row)

    return reduced_matrix


```

```python
def transpose(prematrix): 
    
    postmatrix =[[row[i] for row in prematrix] for i in range(len(prematrix[0]))]
    return postmatrix

#step 2: find smallest element in each column of cost matrix and subtract it from each element in the column

def step_2 (reduced_matrix):

    transposed_matrix = transpose(reduced_matrix)
    
    temp_matrix = step_1(transposed_matrix)
    
    intermediate_matrix = transpose(temp_matrix)

    return intermediate_matrix
```

```python
def single_value(input_row, minimum_values, i, crossed_off_values): 
    
# this method determins whether there is a single value from minimum_values in the row 

    row_zeros = []

    for j in range(len(input_row)):

            # if it is a 0 and it is not in the crossed off values, add it to the list

        if input_row[j] in minimum_values and [i,j] not in crossed_off_values:
            row_zeros.append([i,j])
    
    values = row_zeros[0]
                    
    if len(row_zeros) == 1: 
        return True, values
    else: 
        return False, values
```

```python
def cross_off_coord(intermediate_matrix, crossed_off_values, values, assigned_values, minimum_values): 
    
    p_coord = values[0]
    j_coord = values[1]
   
    
    for l in range(len(intermediate_matrix[p_coord])): 
        
        if intermediate_matrix[p_coord][l] in minimum_values and [p_coord,l] not in assigned_values and \
        [p_coord, l] not in crossed_off_values: 
            
            crossed_off_values.append([p_coord, l])
            
    
    for i in range(len(intermediate_matrix)): 
            
        row = intermediate_matrix[i]
        
        for k in range(len(row)):
        
            #for k in range(len(intermediate_matrix)): 
                
            if intermediate_matrix[k][j_coord] in minimum_values and [k, j_coord] not in assigned_values and \
            [k, j_coord] not in crossed_off_values: 
                            
                crossed_off_values.append([k, j_coord])
                
```

```python
def next_minimum(intermediate_matrix, minimum_values): 
    
    next_min = 10000000000

    for i in range(len(intermediate_matrix)):
        for k in range(len(intermediate_matrix)):

            if intermediate_matrix[i][k]< next_min and intermediate_matrix[i][k] not in minimum_values: 
                next_min = intermediate_matrix[i][k]
    
    minimum_values.append(next_min)
    
    return minimum_values
```
